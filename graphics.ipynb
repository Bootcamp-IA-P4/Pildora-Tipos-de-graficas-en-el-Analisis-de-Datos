```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# 1. Matplotlib (Gráficos básicos)

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Crear un gráfico de líneas
plt.figure(figsize=(8, 6)) # Tamaño del gráfico
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='Datos')  # Personalización
plt.title('Gráfico de Líneas con Matplotlib')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)  # Añadir una cuadrícula
plt.show()

# Crear un gráfico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [10, 15, 7, 12]

plt.figure(figsize=(8, 6))
plt.bar(categorias, valores, color='green')
plt.title('Gráfico de Barras con Matplotlib')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()

# 2. Seaborn (Gráficos estadísticos más atractivos)

# Cargar un conjunto de datos de ejemplo (iris)
iris = sns.load_dataset('iris')

# Crear un diagrama de dispersión (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Diagrama de Dispersión con Seaborn')
plt.show()

# Crear un histograma
plt.figure(figsize=(8, 6))
sns.histplot(iris['sepal_length'], kde=True) # kde es para la línea de densidad
plt.title('Histograma con Seaborn')
plt.xlabel('Longitud del Sépalo')
plt.show()

# Crear un diagrama de cajas (boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Diagrama de Cajas con Seaborn')
plt.show()

# 3. Plotly (Gráficos interactivos)

# Crear un diagrama de dispersión interactivo
fig = px.scatter(iris, x='sepal_length', y='sepal_width', color='species',
                 title='Diagrama de Dispersión Interactivo con Plotly')
fig.show()

# Crear un histograma interactivo
fig = px.histogram(iris, x='sepal_length', color='species',
                   title='Histograma Interactivo con Plotly')
fig.show()

# 4. Scikit-learn (Visualización después de reducción de dimensionalidad)

# Cargar el conjunto de datos Iris desde scikit-learn
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Aplicar PCA (Análisis de Componentes Principales) para reducir a 2 dimensiones
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Crear un DataFrame para facilitar la visualización con Pandas y Seaborn
df = pd.DataFrame(X_reduced, columns=['Componente 1', 'Componente 2'])
df['target'] = y
df['species'] = [target_names[i] for i in y]

# Visualizar los datos reducidos con Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Componente 1', y='Componente 2', hue='species', data=df)
plt.title('PCA: Iris Dataset (2 Componentes)')
plt.show()
```
